# Import

Im ersten Teil des Feature Engineerings wird die grundlegende Struktur des vorliegenden Datensatzes sorgfältig überprüft, einschließlich der Anzahl der Zeilen und Spalten, der NaN-Werte und der Datentypen. Es ist entscheidend, ein erstes Verständnis für den Umfang der behandelten Daten zu haben.

```python
import pandas as pd #Datenanalyse und -manipulation
```

```python
cars_df = pd.read_csv("German_Cars_Data.csv", sep=";")
cars_df  = cars_df .drop("Unnamed: 0", axis= 1) #Unbekannte Modelle fallen lassen
cars_df = cars_df.copy()
```

```python
cars_df.head()
```

```python
cars_df.info()
```

```python
# die features und ihren Datentyp verstehen, NaN prüfen

info_df = pd.DataFrame({
    'Column': cars_df.columns,
    'Non-Null Count': cars_df.count().values,
    'Dtype': cars_df.dtypes.values
})


info_df
```

# Feature Reduktion

Im folgenden werden wir unnötige Features löschen. Dies sind vor allem Features, die zu viele NaN-Werte haben, im Falle eine Vorhersage nicht vorhanden wären oder zu viele unterschiedliche Ausprägungen haben (Individualität).

```python
cars_df["Karosserieform"].value_counts()
```

```python
info_df[["Column", "Non-Null Count"]].sort_values(by="Non-Null Count")
```

```python
interessante_features = ["url",
"Date",
"Country",
"Marke",
"Modell",
"Ort",
"Preis",
"Fahrzeugart",
"Karosserieform",
"Leistung",
"Kilometerstand",
"Fahrleistung p.a.",
"Getriebe",
"Türen",
"Außenfarbe",
"Komfort",
"Sicherheit",
"Extras",
"Sitzplätze",
"Erstzulassung",
"Unterhaltung/Media",
"Hubraum",
"Kraftstoff",
"Innenausstattung",
"Farbe der Innenausstattung",
"Umweltplakette",
"Schadstoffklasse",
"Fahrzeughalter",
"Nichtraucherfahrzeug",
"Scheckheftgepflegt",
"Kraftstoffverbrauch2",
"CO₂-Emissionen2",
"Antriebsart",
"Leergewicht",
"Zylinder",
"Länderversion",
"Gänge",
"Garantie",
"Energieeffizienzklasse",
"Baujahr",
"CO₂-Emissionen (WLTP)2",
"Kraftstoffverbrauch (WLTP)2",
"Letzte Inspektion",
"Elektrische Reichweite (WLTP)",
"Fahrzeugzustand",
"Taxi oder Mietwagen"]
```

```python
# die interessanten Variablen auswählen 
# und einen Datenrahmen nur mit ihnen erstellen
cars_df = cars_df[interessante_features]
```

```python
cars_df.info()
```

```python
cars_df.to_csv('Zwischenspeicher_Cars_Data.csv', index=True,  sep=";")
```

Ideen:

- Erstzulassung anstatt Baujahr
- Bei Sicherheit, Komfort etc. mit Anzahl an Buchstaben in eine Skala einteilen
- Bei Fahrzeugzustand, für alle anderen Fahrzeuge nicht Unfallfahrzeug, fahrtauglich angeben
